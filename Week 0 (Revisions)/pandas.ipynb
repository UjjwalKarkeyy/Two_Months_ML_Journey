# Concat and Merging

```python
df1 = pd.DataFrame({
    'A' : ['A0','A1','A2','A3'],
    'B' : ['B0','B1','B2','B3'],
    'C' : ['C0','C1','C2','C3'],
    'D' : ['D0','D1','D2','D3'],
}, index = [0,1,2,3])
```

```python
df2 = pd.DataFrame({
    'A' : ['A4','A5','A6','A7'],
    'B' : ['B4','B5','B6','B7'],
    'C' : ['C4','C5','C6','C7'],
    'D' : ['D4','D5','D6','D7'],
}, index = [4,5,6,7])
```

```python
df1
```

```python
df2
```

```python
results = pd.concat([df1, df2])
```

```python
results
```

```python
df1 = pd.DataFrame({
    'A' : ['A0','A1','A2','A3'],
    'B' : ['B0','B1','B2','B3'],
    'C' : ['C0','C1','C2','C3'],
    'D' : ['D0','D1','D2','D3'],
})

df2 = pd.DataFrame({
    'E' : ['A4','A5','A6','A7'],
    'F' : ['B4','B5','B6','B7'],
    'C' : ['C0','C3','C6','C7'],
    'G' : ['D4','D5','D6','D7'],
})   
```

```python
df1
```

```python
df2
```

```python
# merging

inner_df = pd.merge(df1, df2, on="C", how="inner")
left_df = pd.merge(df1, df2, on="C", how="left")
```

```python
inner_df
left_df
```

# loc vs iloc

The main difference is that:

- loc gets rows (and/or columns) with particular labels.
- iloc gets rows (and/or columns) at integer locations.

```python
# loc vs iloc

s = pd.Series(list('abcd'), index = [1,2,0,4])
```

```python
s
```

```python
print(s.loc[0]) # accessing at 0 index
s.iloc[0]
```

```python
print(s.loc[1:2]) # loc is inclusive
print(s.iloc[1:2]) # iloc is exclusive
```
